```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)


```

```python
city = cities[0]

###inputs for URL 
url = "http://api.openweathermap.org/data/2.5/weather?"

#### Set unit to Imperial to obtain Temp (F) and windspeed (mph)
units = "imperial"
##load the api_key for OpenWeatherMap
api_key = weather_api_key
    
# Build query URL and request your results in Celsius
query_url = f"{url}appid={api_key}&q={city}&units={units}"

# Get weather data
weather_response = requests.get(query_url)
weather_json = weather_response.json()

weather_json
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.

city_retrieved = []
lat = []
lon = []
temperature = []
humidity = []
cloud = []
wind_speed = []
date_capture = []
country = []

###inputs for URL 
url = "http://api.openweathermap.org/data/2.5/weather?"

#### Set unit to Imperial to obtain Temp (F) and windspeed (mph)
units = "imperial"

##load the api_key for OpenWeatherMap
api_key = weather_api_key

# Capture Data for cities

cityweather_df= []


########Looping inputs

##set counter
counter_set = 1
counter_record = 1

###information to print

print("Beginning Data Retrieval")
print("-----------------------------")


for rnd_city in cities:

    city = rnd_city #cities[0]
    
    # Build query URL and request your results in Fahrenheit 
    query_url = f"{url}appid={api_key}&q={city}&units={units}"

    # Get weather data
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()

    
    # Try to grab the height and mass of characters if they are available in the Star Wars API
    try:
        #Append metrics for scatter plots

        #Latitude
        #weather_json['coord']['lat']
        lat.append(weather_json['coord']['lat'])

        #Longitude
        #weather_json['coord']['lat']
        lon.append(weather_json['coord']['lon'])

        #Temperature (F)
        temperature.append(weather_json['main']['temp_max'])

        #Humidity (%)
        humidity.append(weather_json['main']['humidity'])

        #Cloudiness (%)
        cloud.append(weather_json['clouds']['all'])

        #Wind Speed (mph)
        wind_speed.append(weather_json['wind']['speed'])

        #Date
        date_capture.append(weather_json['dt'])

        ##Country
        country.append(weather_json['sys']['country'])
                
        #city
        city_retrieved.append(city)
        
        print(f"Processing Record {counter_record} of Set {counter_set} | {rnd_city}")
        
        if (counter_record <50):
            counter_record += 1
        else:
            counter_record = 0
            counter_set += 1
        
    # Handle exceptions for instances where the city is not found in the Weather data
    except:
        # Append null values
        print("City not found. Skipping...")
        pass
    
# Print completion of data load 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")

```

```python
print(f"{len(city_retrieved)}, {len(lat)}, {len(lon)}, {len(temperature)}, {len(humidity)}, {len(cloud)}, {len(wind_speed)}, {len(country)}, {len(date_capture)}")
```
