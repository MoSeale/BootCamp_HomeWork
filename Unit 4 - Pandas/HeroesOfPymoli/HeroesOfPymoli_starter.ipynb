## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items in "Item ID" column of purchase_data
unique_items=len(purchase_data["Item ID"].unique())
unique_items

# Total Revenue sum of "Price" column of purchase_data
sum_price=purchase_data["Price"].sum()
sum_price

#Number of Purchases using "Purchase ID" column of purchase_data
num_purchases= len(purchase_data["Purchase ID"].unique())
num_purchases

# Average Price of items in "Price" column of purchase_data
avg_price= sum_price/num_purchases
avg_price

#Purchasing Analysis Summary
pa_summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Purchase Price": [avg_price],
                              "Number of Purchases": [num_purchases],
                              "Total Revenue": [sum_price]
                             })

pa_summary_df["Average Purchase Price"] = pa_summary_df["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)
pa_summary_df["Total Revenue"] = pa_summary_df["Total Revenue"].astype(float).map(
    "${:,.2f}".format)

pa_summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
##create a subset of the purchase_data
columns = ["Item ID", "Item Name","Price"]

popular_df=purchase_data[columns]

popular_df.head()

# Group by Item ID and Item Name. 
#Perform calculations to obtain purchase count, average item price, and total purchase value

group_popular_df=popular_df.groupby(["Item ID", "Item Name"]).agg({"Item ID": 'count', "Price": ['mean']}) 

group_popular_df.head()

#Column Rename
group_popular_df2=group_popular_df.rename(
    columns= {'Item ID':"Purchase Count",
              'count':"",
              "Price":"Item Price",
              'mean':""          
             })
group_popular_df2

#Generate Total Purchase Value (separately since PRice was repeated)
Total_Purchase_Value=popular_df.groupby(["Item ID", "Item Name"]).agg({"Price": ['sum']})
Total_Purchase_Value
Total_Purchase_Value["Price"]['sum']

group_popular_df2["Total Purchase Value"]=Total_Purchase_Value["Price"]['sum']
group_popular_df2

# Sort the DataFrame by the values in the "Purchase Count" column to find the worst
group_popular_df3 = group_popular_df2.sort_values(by ="Purchase Count" , ascending = False)


#Formatting
final_group_popular_df =group_popular_df3
final_group_popular_df["Item Price"] = final_group_popular_df["Item Price"].astype(float).map("${:,.2f}".format)
final_group_popular_df["Total Purchase Value"] = final_group_popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)


final_group_popular_df.head(5)


```
